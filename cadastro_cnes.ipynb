## Cadastro Nacional de Estabelecimentos de Saúde - CNES

<p>Localização dos estabelecimentos registrados no Cadastro Nacional de Estabelecimentos de Saúde. Sobre o CNES, ver: http://cnes.datasus.gov.br/ Mais informações sobre licença de uso em: http://bvsms.saude.gov.br/bvs/saudelegis/gm/2015/prt164 6_02_10_2015.html Excetuando-se os casos previstos na Lei nº 12.527, de 18 de novembro de 2011, as informações existentes no CNES são públicase de amplo acesso.</p>
<p>A coordenada de cada estabelecimento é obtida do próprio CNES e validada por meio de operações espaciais. Essas operaçõesverificam se o ponto está no município, considerando-se um raio de 5.000 metros. Quando a coordenada não está correta, sãofeitas buscas em outros sistemas do Ministério da Saúde e em serviços web como o Google Maps. Por fim, caso a coordenadanão tenha sido obtida de forma correta nesse processo, utiliza-se a coordenada da sede do município.</p> 
<p>A fonte final éregistrada no banco de dados em uma coluna específica.Periodicamente as coordenadas são revisadas com o objetivo de melhorar a qualidade do dado, evitando-se ao máximo o usoda sede do município como a opção para determinação da coordenada.</p>

```python
coord = df[['lat','long']]
coord.head()
```

```python
lat = df['lat'][:1000].values
long = df['long'][:1000].values
```

```python
mapa = fl.Map(location=[-15.788497,-47.879873],zoom_start= 5)
for la,lo in zip(lat,long):
    fl.Marker([la,lo]).add_to(mapa)
mapa
```
